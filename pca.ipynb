```python
from PCA_tool import PCA_tool
import pandas as pd
import numpy as np
```

```python
# load data
data_x = pd.read_excel('sample_data.xlsx', sheet_name='X', index_col=0)
data_y = pd.read_excel('sample_data.xlsx', sheet_name='y2', index_col=0)
```

```python
# initialize pca tool
pca_tool = PCA_tool(df_x = data_x, df_y=data_y)

# perform pca
pca_tool.perform_pca(scale=True)
```

```python
# save all necessary data if needed

pca_tool.save_data('my_pca_results')
```

```python
# scree plot

pca_tool.plot_scree(save=True, num_pcs=10)
```

```python
# loadings plot

pca_tool.plot_loadings(n_pcs=2, thresh=0.7, save=True)
```

```python
# generate scatter - use .plot_scatter() method 
# (without any args generates default scatter for PC1 v. PC2)
# use below to tune your plot

pca_tool.plot_scatter(

            # which pcs to plot
            plot_pcs=[1,2],

            # add annotations and set their relative x,y positions
            annot=False,
            annot_pos=[0.1, 0.35], 

            # markers size
            msize=200,

            # color by y
            color_by_y=True,

            # use predefined matplotlib colorscale
            colorscale='RdBu',

            # opacity
            alpha = 1,

            # markers edge colors
            edgecolors='k',


            # play with figsize (x, y)
            figsize = (12,11),

            # save figure
            save = True,
            )
```
